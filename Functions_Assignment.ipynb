1. What is the difference between a function and a method in Python?

 - A function is a block of code that performs a specific task and is defined independently.
 - A method is a function that is associated with an object or class.

```python
#Example
# Function
def add(a, b):
    return a + b

print(add(2, 3))

# Method
text = "hello"
print(text.upper())
```

2. Explain the concept of function arguments and parameters in Python.

 - Parameters are variables defined in the function definition.
 - Arguments are values passed to the function when calling it.

```python
#Example
def greet(name):   # name is parameter
    print("Hello", name)

greet("Alice")     # "Alice" is argument
```

3. What are the different ways to define and call a function in Python?

 - Functions can be defined using the def keyword and called by their name.


```python
#Example
# Normal function
def square(n):
    return n * n

print(square(4))

# Lambda function
square = lambda x: x * x
print(square(5))
```

4. What is the purpose of the return statement in a Python function?

 - The return statement sends a value back to the function caller and ends the function execution.

```python
#Example
def multiply(a, b):
    return a * b

result = multiply(3, 4)
print(result)
```

5. What are iterators in Python and how do they differ from iterables?

 - An iterable is an object that can return an iterator.
 - An iterator is an object that produces values one at a time using __next__().

```python
#Example
lst = [1, 2, 3]      # Iterable
it = iter(lst)       # Iterator

print(next(it))
print(next(it))
```

6. Explain the concept of generators in Python and how they are defined.

 - Generators are functions that return values one at a time using the yield keyword.

```python
#Example
def count_up():
    yield 1
    yield 2
    yield 3

for num in count_up():
    print(num)
```

7. What are the advantages of using generators over regular functions?
 - Saves memory
 - Produces values lazily
 - Efficient for large data

```python
#Example
def gen():
    yield 10
    yield 20

print(list(gen()))
```

8. What is a lambda function in Python and when is it typically used?

 - A lambda function is a small anonymous function used for short operations.

```python
#Example
add = lambda a, b: a + b
print(add(3, 5))
```

9. Explain the purpose and usage of the map() function in Python.

 - The map() function applies a function to each element of an iterable.

```python
#Example
nums = [1, 2, 3, 4]
result = list(map(lambda x: x * 2, nums))
print(result)
```

10. What is the difference between map(), reduce(), and filter() functions in Python?
 - map() → transforms data
 - filter() → selects data
 - reduce() → combines data into one value

```python
#Examlpe
from functools import reduce

nums = [1, 2, 3, 4]

print(list(map(lambda x: x * 2, nums)))
print(list(filter(lambda x: x % 2 == 0, nums)))
print(reduce(lambda a, b: a + b, nums))
```

11. Internal mechanism of sum operation using reduce() on list [47, 11, 42, 13]

 - Using reduce() step-by-step:

   - Step 1: 47 + 11 = 58
   - Step 2: 58 + 42 = 100
   - Step 3: 100 + 13 = 113
   -  Final Result = 113

```python
#Example
from functools import reduce

nums = [47, 11, 42, 13]
result = reduce(lambda a, b: a + b, nums)
print(result)
```

```python
# 1. Function to return sum of all even numbers in a list
def sum_even(numbers):
    return sum(num for num in numbers if num % 2 == 0)

print(sum_even([1, 2, 3, 4, 5, 6]))


# 2. Function to reverse a string
def reverse_string(s):
    return s[::-1]

print(reverse_string("Python"))


# 3. Function to return squares of numbers in a list
def square_list(numbers):
    return [num ** 2 for num in numbers]

print(square_list([1, 2, 3, 4]))


# 4. Function to check if a number is prime (1 to 200)
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

print(is_prime(29))


# 5. Iterator class for Fibonacci sequence
class Fibonacci:
    def __init__(self, terms):
        self.terms = terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration
        self.count += 1
        self.a, self.b = self.b, self.a + self.b
        return self.a

fib = Fibonacci(5)
for num in fib:
    print(num)


# 6. Generator to yield powers of 2
def powers_of_two(exp):
    for i in range(exp + 1):
        yield 2 ** i

for val in powers_of_two(5):
    print(val)


# 7. Generator to read file line by line
def read_file(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()

# Example usage (file must exist)
# for line in read_file("sample.txt"):
#     print(line)


# 8. Lambda to sort list of tuples by second element
data = [(1, 3), (4, 1), (2, 2)]
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)


# 9. Using map() to convert Celsius to Fahrenheit
celsius = [0, 20, 30, 40]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print(fahrenheit)


# 10. Using filter() to remove vowels from a string
string = "Hello World"
result = ''.join(filter(lambda x: x.lower() not in 'aeiou', string))
print(result)


# 11. Book shop accounting problem using lambda and map
orders = [
    (34587, "Learning Python, Mark Lutz", 4, 40.95),
    (98762, "Programming Python, Mark Lutz", 5, 56.80),
    (77226, "Head First Python, Paul Barry", 3, 32.95),
    (88112, "Einführung in Python3, Bernd Klein", 3, 24.99)
]

result = list(map(
    lambda order: (
        order[0],
        order[2] * order[3] + 10 if order[2] * order[3] < 100 else order[2] * order[3]
    ),
    orders
))

print(result)
```
